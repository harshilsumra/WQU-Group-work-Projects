Q 8.1 & 8.2 Common and difference in PCA and lasso
<br>Ans: PCA can be used as a dimensionality reduction technique if you drop Principal Components based on a heuristic, but it offers no feature selection, as the Principal Components are retained instead of the original features. However, tuning the number of Principal Components retained should work better than using heuristics, unless there are many low variance components and you are simply interested in filtering them.

<br> LASSO (ℓ1 regularization) on the other hand can, intrinsically, perform feature selection as the coefficients of predictors are shrunk towards zero. It still requires hyperparameter tuning because there's a regularization coefficient that weights how severe is the regularization of the loss function.

```python
X = data.drop('LUXXX', axis = 1) ## controlling variable)
y = data.LUXXX ## target Variable
```

```python
#  we fitted  the Linear regression using Principal Component Analysis using 10 features extracted 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

# Principal components regression (PCA)
steps = [
    ('scale', StandardScaler()),
    ('pca', PCA()),
    ('estimator', LinearRegression())
]
pipe = Pipeline(steps)
pca = pipe.set_params(pca__n_components=10)
pca.fit(X, y)
```

```python
Pipeline(memory=None,
         steps=[('scale',
                 StandardScaler(copy=True, with_mean=True, with_std=True)),
                ('pca',
                 PCA(copy=True, iterated_power='auto', n_components=10,
                     random_state=None, svd_solver='auto', tol=0.0,
                     whiten=False)),
                ('estimator',
                 LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None,
                                  normalize=False))],
         verbose=False)
```

```python
### Checking for the Regression Performance 
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, r2_score
y_pred = pca.predict(X)
rsquared_c = r2_score(y, y_pred)
mse_c = mean_squared_error(y, y_pred)
print(f"The R-squared error for the PCA regression model is {rsquared_c}")
print(f"The Mean_Squared_Error for the PCA regression model is {mse_c}")
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from scipy.stats import uniform
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")

#Q 10.3 Split the data into testing and training, with 2/3 for training and 1/3 for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=2/3, test_size=1/3)
alpha_lasso = [(i+1)/1000 for i in range(1000)]
alpha_num = 1000
pred_num = X.shape[1]
coeff_a = np.zeros((alpha_num, pred_num))
train_r_squared = np.zeros(alpha_num)
test_r_squared = np.zeros(alpha_num)

for i in range(1000):    
    reg = Lasso(alpha = alpha_lasso[i])
    reg.fit(X_train, y_train)
    coeff_a[i,:] = reg.coef_
    train_r_squared[i] = reg.score(X_train, y_train)
    test_r_squared[i] = reg.score(X_test, y_test)
    
#prin the( train_r_squared)
plt.plot(train_r_squared, 'bo-', label=r'$R^2$ Training set', color="darkgreen", alpha=0.6, linewidth=3)
plt.plot(test_r_squared, 'bo-', label=r'$R^2$ Test set', color="darkred", alpha=0.6, linewidth=3)
plt.xlabel('alpha index'); plt.ylabel(r'$R^2$')
plt.title(r'Evaluate lasso regression with 1000 alphas')
plt.legend(loc='best')
plt.grid()
```

```python
df_lam = pd.DataFrame(test_r_squared, columns=['R_squared'])
df_lam['alpha'] = (alpha_lasso)
# Checking to return the index of the row where column has maximum value.
df_lam.loc[df_lam['R_squared'].idxmax()]
```

```python
#Checking for the Coefficients of best model to ascertan presictors 
reg_best = Lasso(alpha=1)
reg_best.fit(X_train, y_train)
print("The best model coefficients")
print(reg_best.coef_)
print("Mean squared error (MSE) of the best model on the test data")

print(mean_squared_error(y_test, reg_best.predict(X_test)))
```

```python
### Checking out the performance of the best features  of the lasso model 

coef = pd.Series(reg_best.coef_, index = X.columns)
imp_coef = coef.sort_values()
import matplotlib
matplotlib.rcParams['figure.figsize'] = (8.0, 10.0)
imp_coef.plot(kind = "barh")
plt.title(" generating Feature importance using Lasso Model")
```
